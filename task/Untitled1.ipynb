```python
import matplotlib.pyplot as plt
import numpy as np

# Generating sample data for transmission functions of two bands
wavelengths = np.linspace(300, 800, 500)  # Wavelengths ranging from 300nm to 800nm
transmission_band1 = np.exp(-0.5 * ((wavelengths - 450) / 40) ** 2)  # Transmission function for band 1
transmission_band2 = np.exp(-0.5 * ((wavelengths - 550) / 40) ** 2)  # Transmission function for band 2

# Plotting the transmission functions
plt.figure(figsize=(10, 5))
plt.plot(wavelengths, transmission_band1, label='Band 1 Transmission Function')
plt.plot(wavelengths, transmission_band2, label='Band 2 Transmission Function')
plt.fill_between(wavelengths, transmission_band1, transmission_band2, color='gray', alpha=0.3)
plt.title('Overlap of Transmission Functions for Two Different Bands')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Transmission')
plt.legend()
plt.grid(True)
plt.show()

```
