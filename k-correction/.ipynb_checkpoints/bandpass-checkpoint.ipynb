```python
# Replace this with the path to your file if it's not in the same directory
file_path = 'Palomar_ZTF.r.dat'

# Initialize two empty lists to store wavelengths and values separately
wavelengths = []
values = []

# Open the file and read the lines
with open(file_path, 'r') as file:
    lines = file.readlines()

# Process the lines as needed
for line in lines:
    # Split the line into parts, for example, by space
    parts = line.strip().split(' ')
    # Append the wavelength and value to their respective lists
    if len(parts) == 2:  # Ensure that there are exactly two parts
        wavelengths.append(float(parts[0]))
        values.append(float(parts[1]))

# Combine the two lists
g_band_bandpass = [wavelengths, values]
```

```python
from astropy.io import fits

# Replace this with the path to your FITS file
file_path = 'desi-ztf-qso-iron-1000-20210517.fits'

# Open the FITS file
with fits.open(file_path) as hdul:
    # Access the HDU1
    hdu1 = hdul[1]  # HDU indices start from 0, so 1 corresponds to HDU1

    # Read the image data
    wavelength = hdu1.data
```

```python
import numpy as np

# Function to find the nearest index in the wavelength array
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx

# New array to store the results
new_g_band_bandpass = [[], []]

# Iterate over the wavelength array and interpolate values
for wl in wavelength:
    if g_band_bandpass[0][0] <= wl <= g_band_bandpass[0][-1]:
        if wl in g_band_bandpass[0]:
            # Exact match found
            idx = g_band_bandpass[0].index(wl)
            bandpass_value = g_band_bandpass[1][idx]
        else:
            # Interpolate
            idx = find_nearest(g_band_bandpass[0], wl)
            # Find the nearest indices to the left and right
            if wl > g_band_bandpass[0][idx]:
                left = idx
                right = min(idx + 1, len(g_band_bandpass[0]) - 1)
            else:
                right = idx
                left = max(idx - 1, 0)

            # Calculate the interpolated value
            wl_left = g_band_bandpass[0][left]
            wl_right = g_band_bandpass[0][right]
            bandpass_left = g_band_bandpass[1][left]
            bandpass_right = g_band_bandpass[1][right]

            # Weighted interpolation
            bandpass_value = ((wl - wl_left) / (wl_right - wl_left) * bandpass_right +
                              (wl_right - wl) / (wl_right - wl_left) * bandpass_left)

        # Add the results to the new array
        new_g_band_bandpass[0].append(wl)
        new_g_band_bandpass[1].append(bandpass_value)

# new_g_band_bandpass now contains wavelengths and their corresponding bandpass values
```
